```python
saved_model_path='D:\natasha\kuliah\sem 6\bangkit\coba python\project capstone\hasil model'
model = tf.keras.models.load_model('/content/drive/MyDrive/mymodelhopefullyfinal.h5')
```

```python
converter = tf.lite.TFLiteConverter.from_keras_model(model)

# Set the optimzations
converter.optimizations = [tf.lite.OpsSet.TFLITE_BUILTINS,
                           tf.lite.OpsSet.SELECT_TF_OPS]

# Invoke the converter to finally generate the TFLite model
tflite_model = converter.convert()
```

```python
import pathlib

tflite_model_file = pathlib.Path('/content/drive/MyDrive/tflite/model.tflite')
tflite_model_file.write_bytes(tflite_model)
```
