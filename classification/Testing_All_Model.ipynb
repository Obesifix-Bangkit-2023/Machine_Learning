```python
import tensorflow as tf
import os
```

```python
model_adam = tf.keras.models.load_model("/content/drive/MyDrive/capstone/checkpoint_all/model_prep_adam.h5")
model_rmsprop = tf.keras.models.load_model("/content/drive/MyDrive/capstone/checkpoint_all/model_prep.h5")
model_sgd = tf.keras.models.load_model("/content/drive/MyDrive/capstone/checkpoint_all/mymodelhopefullyfinal.h5")
```

```python
food_name = os.listdir("/content/drive/MyDrive/capstone/dataset fix/train")
food_name.sort()
food_name
```

```python
from google.colab import files
from tensorflow.keras.utils import load_img, img_to_array
import matplotlib.pyplot as plt

uploaded = files.upload()

for fn in uploaded.keys():
  count = 1
  plt.subplot(3,3,count)
  plt.imshow(plt.imread(fn)/255.)
  plt.axis("off")
  plt.show()
  count+=1
  # predicting images
  path = fn
  img = load_img(path, target_size=(256, 256))
  x = img_to_array(img)
  x = tf.expand_dims(x, axis=0)/255.

  classes_adam = model_adam.predict(x, batch_size=128).argmax(axis=1)
  classes_rmsprop = model_rmsprop.predict(x, batch_size=128).argmax(axis=1)
  classes_sgd = model_sgd.predict(x, batch_size=128).argmax(axis=1)
  classes_adam = classes_adam[0]
  classes_rmsprop = classes_rmsprop[0]
  classes_sgd = classes_sgd[0]
  #print(fn)
  print(f"Adam : {food_name[classes_adam]}")
  print(f"RMSprop : {food_name[classes_rmsprop]}")
  print(f"SGD : {food_name[classes_sgd]}")
```
