```python
data = pd.read_csv(r'/content/drive/MyDrive/capstone/recommendation system/recipes.csv')
```

```python
list_fruit = ['Apple', 'Avocados', 'Fruit', 'Cherries', 'Citrus', 'Fruit', 'Mango', 'Melons', 'Oranges', 'Pears', 'Pineapple', 'Strawberry', 'Lemon', 'Pumpkin', 'Tropical Fruits', 'Coconut', 'Desserts Fruit']
list_vegetable = ['Vegetable', 'Greens', 'Cauliflower', 'Corn', 'Potato', 'Spinach', 'Peppers', 'Chard', 'Beans']
list_beef = ['Beef Organ Meats', 'Ham', 'Meatballs', 'Meatloaf', 'Meat', 'Roast Beef', 'Steak', 'Veal']
list_chicken = ['Chicken', 'Chicken Breast', 'Chicken Livers', 'Chicken Thigh & Leg', 'Whole Chicken']
list_duck = ['Whole Duck', 'Duck Breasts']
list_fish = ['Tuna', 'Tilapia', 'Catfish', 'Trout', 'Perch', 'Halibut']
list_pork = ['Pork']
list_lamb = ['Lamb/Sheep']
list_seafood = ['Tuna', 'Crab', 'Lobster', 'Squids', 'Mussels']
list_bread = ['Yeast Breads', 'Sourdough Breads', 'Quick Breads', 'Breads', 'Scones']
list_soup = ['Chowders', 'Clear Soup', 'Mushroom Soup', 'Gumbo']
list_low_chol = ['Low Cholesterol']
list_low_protein = ['Low Protein']
list_hi_protein = ['High Protein']
list_soy = ['Soy/Tofu', 'Tempeh']
list_dairy = ['Cheese', 'Cheesecake', 'Macaroni And Cheese']
list_rice = ['Rice', 'White Rice', 'Long Grain Rice', 'Brown Rice', 'Short Grain Rice']
list_cookies = ['Drop Cookies', 'Chocolate Chip Cookies', 'Bar Cookies']
list_pasta = ['Pasta Shells', 'Spaghetti', 'Pasta', 'Macaroni and Cheese']
list_breakfast = ['Breakfast']
list_dessert = ['Dessert', 'Desserts Fruit', 'Frozen Dessert', 'Ice Cream', 'Jellies']
list_lunch = ['Lunch/Snacks']
```

```python
def change_recipecategory():
  data.insert(1, 'FoodCategory', np.nan)
  for i in list_fruit:
    data.loc[data["RecipeCategory"] == i, "FoodCategory"] = 'Fruit'
  for i in list_vegetable:
    data.loc[data["RecipeCategory"] == i, "FoodCategory"] = 'Vegetable'
  for i in list_beef:
    data.loc[data["RecipeCategory"] == i, "FoodCategory"] = 'Beef'
  for i in list_chicken:
    data.loc[data["RecipeCategory"] == i, "FoodCategory"] = 'Chicken'
  for i in list_duck:
    data.loc[data["RecipeCategory"] == i, "FoodCategory"] = 'Duck'
  for i in list_fish:
    data.loc[data["RecipeCategory"] == i, "FoodCategory"] = 'Fish'
  for i in list_pork:
    data.loc[data["RecipeCategory"] == i, "FoodCategory"] = 'Pork'
  for i in list_lamb:
    data.loc[data["RecipeCategory"] == i, "FoodCategory"] = 'Lamb'
  for i in list_seafood:
    data.loc[data["RecipeCategory"] == i, "FoodCategory"] = 'Seafood'
  for i in list_bread:
    data.loc[data["RecipeCategory"] == i, "FoodCategory"] = 'Bread'
  for i in list_soup:
    data.loc[data["RecipeCategory"] == i, "FoodCategory"] = 'Soup'
  for i in list_low_chol:
    data.loc[data["RecipeCategory"] == i, "FoodCategory"] = 'Low Cholestrol'
  for i in list_low_protein:
    data.loc[data["RecipeCategory"] == i, "FoodCategory"] = 'Low Protein'
  for i in list_hi_protein:
    data.loc[data["RecipeCategory"] == i, "FoodCategory"] = 'High Protein'
  for i in list_soy:
    data.loc[data["RecipeCategory"] == i, "FoodCategory"] = 'Soy Product'
  for i in list_dairy:
    data.loc[data["RecipeCategory"] == i, "FoodCategory"] = 'Dairy Product'
  for i in list_rice:
    data.loc[data["RecipeCategory"] == i, "FoodCategory"] = 'Rice'
  for i in list_cookies:
    data.loc[data["RecipeCategory"] == i, "FoodCategory"] = 'Cookies'
  for i in list_pasta:
    data.loc[data["RecipeCategory"] == i, "FoodCategory"] = 'Pasta'
  for i in list_breakfast:
    data.loc[data["RecipeCategory"] == i, "FoodCategory"] = 'Breakfast'
  for i in list_dessert:
    data.loc[data["RecipeCategory"] == i, "FoodCategory"] = 'Dessert'
  for i in list_lunch:
    data.loc[data["RecipeCategory"] == i, "FoodCategory"] = 'Lunch'
  
  return data
```
