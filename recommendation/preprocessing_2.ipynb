```python
import pandas as pd
import numpy as np
```

Reading the file

```python
data = pd.read_csv(r'/content/drive/MyDrive/capstone/recommendation system/new_recommendation.csv')
```

Change the calories etc to per portion

```python
data['Calories']=data['Calories']/data['RecipeServings']
data['FatContent']=data['FatContent']/data['RecipeServings']
data['SaturatedFatContent']=data['SaturatedFatContent']/data['RecipeServings']
data['CholesterolContent']=data['CholesterolContent']/data['RecipeServings']
data['SodiumContent']=data['SodiumContent']/data['RecipeServings']
data['CarbohydrateContent']=data['CarbohydrateContent']/data['RecipeServings']
data['FiberContent']=data['FiberContent']/data['RecipeServings']
data['SugarContent']=data['SugarContent']/data['RecipeServings']
data['ProteinContent']=data['ProteinContent']/data['RecipeServings']
```

Sort by values

```python
data=data.sort_values(by=['Calories'])
```

#newcolumns for calorie

```python
calories_standart=[]
for row in data['Calories']:
        if row >= 1000 :    calories_standart.append('high_cal')
        elif row < 1000 and row >= 650:   calories_standart.append('midhigh_cal')
        elif row < 650 and row >= 325:  calories_standart.append('so_so_cal')
        elif row < 325:  calories_standart.append('low_cal')
```

```python
data['cal_standart'] = calories_standart
```

```python
data
```

# for fat

```python
fat_standart=[]
for row in data['FatContent']:
        if row >= 17.5 :    fat_standart.append('high_fat')
        elif row < 17.5 and row >=3 :  fat_standart.append('so_so_fat')
        elif row < 3:  fat_standart.append('low_fat')
```

```python
data['fat_standart'] = fat_standart
```

```python
data
```

#for carb

```python
carbo_standart=[]
for row in data['CarbohydrateContent']:
        if row >= 75 :    carbo_standart.append('high_carb')
        elif row < 75 and row >= 45:  carbo_standart.append('so_carb')
        elif row < 45:  carbo_standart.append('low_carb')
```

```python
data['carbo_standart'] = carbo_standart
```

```python
data
```

#for protein

```python
protein_standart=[]
for row in data['ProteinContent']:
        if row >= 56 :    protein_standart.append('high_pro')
        elif row < 56 and row >= 46:  protein_standart.append('so_so_pro')
        elif row < 46:  protein_standart.append('low_pro')
```

```python
data['protein_standart'] = protein_standart
```
